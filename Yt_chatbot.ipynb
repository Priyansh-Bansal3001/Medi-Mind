```python
from youtube_transcript_api import YouTubeTranscriptApi, TranscriptsDisabled
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings, ChatHuggingFace,HuggingFaceEndpoint
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

```python
video_id = 'JAgHUDhaTU0'
ytt_api = YouTubeTranscriptApi()
try:
  transcript_list = ytt_api.fetch(video_id ,languages=["en"])

  transcript = " ".join(chunk.text for chunk in transcript_list)
  print(transcript)

except TranscriptsDisabled:
  print("No caption available for this video")
```

```python
transcript_list
```

```python
splitter  = RecursiveCharacterTextSplitter(chunk_size = 500 , chunk_overlap = 100)
chunks = splitter.create_documents([transcript])
```

```python
chunks[0]
```

```python
embeddings = HuggingFaceEmbeddings(model = "Qwen/Qwen3-Embedding-0.6B")
vector_store = FAISS.from_documents(chunks , embeddings)
```

```python
vector_store = FAISS.from_documents(chunks , embeddings)
```

```python
vector_store.index_to_docstore_id
```

```python
from re import search
retriever = vector_store.as_retriever(search_type = "similarity", search_kwargs = {"k" : 4})
```

```python
retriever
```

```python
retriever.invoke("what is open source")
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"  # your HF model name

# Load tokenizer & model
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")

# Create a text-generation pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    temperature=0.2
)

hf_pipeline = HuggingFacePipeline(pipeline=pipe)

# Now pass into ChatHuggingFace
llm = ChatHuggingFace(llm=hf_pipeline)
```

```python
prompt = PromptTemplate(
    template = """
    You are a helpful assisstant.
    Answer ONLY form the provided transcript context.
    if the context is insufficient, just say you dont know.

    {context}
    Question : {question}
    """,
    input_variables = {'context','question'}
)
```

```python
question = "Who is the speaker in this video"
retrieved_docs = retriever.invoke(question)
```

```python
context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
```

```python
final_prompt = prompt.invoke({"context" :  context_text,"question" : question})
```

```python
from langchain.schema.runnable import RunnableParallel, RunnableLambda , RunnablePassthrough
from langchain.schema import StrOutputParser
```

```python
def format_docs(retrieved_docs):
  context_text = "\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

```python
parallel_chain = RunnableParallel({
   'context' : retriever | RunnableLambda(format_docs),
   'question' : RunnablePassthrough()
})
```

```python
parallel_chain.invoke('Who is the speaker in this video')
```

```python
from langchain.schema import BaseOutputParser

class AssistantOutputParser(BaseOutputParser[str]):
    def parse(self, text: str) -> str:
        # HuggingFace models often output: <|user|> ... </s><|assistant|> answer ...
        if "<|assistant|>" in text:
            text = text.split("<|assistant|>")[-1]
        # Remove </s> or any trailing tags
        text = text.replace("</s>", "").strip()
        return text

parser = AssistantOutputParser()
```

```python
main_chain = parallel_chain | prompt | llm | parser
```

```python
import gradio as gr
from youtube_transcript_api import YouTubeTranscriptApi

# 🔹 Your LangChain pipeline
# (assuming main_chain = parallel_chain | prompt | llm | parser is already built)
ytt_api = YouTubeTranscriptApi()
def answer_question(video_id, question):
    try:
        # fetch transcript
        transcript = ytt_api.fetch(video_id ,languages = ['en'])
        transcript_text = " ".join(chunk.text for chunk in transcript)

        # pass both context & question to chain
        result = main_chain.invoke(question)

        return str(result)

    except Exception as e:
        return f"❌ Error: {e}"

# --- UI with Gradio ---
with gr.Blocks() as demo:
    gr.Markdown("## 🎥 YouTube Transcript Q&A Bot")
    video_id = gr.Textbox(label="Enter YouTube Video ID (e.g., dQw4w9WgXcQ)")
    question = gr.Textbox(label="Ask your question about the video")
    answer = gr.Textbox(label="Answer")

    btn = gr.Button("Get Answer")
    btn.click(fn=answer_question, inputs=[video_id, question], outputs=answer)

demo.launch()
```
